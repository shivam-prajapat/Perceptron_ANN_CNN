```python
#importing dependencies
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.utils import to_categorical
```

```python
#Load & Preprocess Data
(X_train, y_train), (X_test, y_test) = mnist.load_data()
X_train = X_train / 255.0
X_test = X_test / 255.0
y_train_cat = to_categorical(y_train, 10)
y_test_cat = to_categorical(y_test, 10)
```

```python
perceptron = Sequential([
Flatten(input_shape=(28,28)),
Dense(10, activation='softmax')
])
perceptron.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
h1 = perceptron.fit(X_train, y_train_cat, epochs=5, validation_data=(X_test, y_test_cat))
```

```python
ann = Sequential([
Flatten(input_shape=(28,28)),
Dense(128, activation='relu'),
Dense(10, activation='softmax')
])
ann.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
h2 = ann.fit(X_train, y_train_cat, epochs=5, validation_data=(X_test, y_test_cat))
```

```python
X_train_cnn = X_train.reshape(-1,28,28,1)
X_test_cnn = X_test.reshape(-1,28,28,1)
cnn = Sequential([
Conv2D(32, (3,3), activation='relu', input_shape=(28,28,1)),
MaxPooling2D((2,2)),
Conv2D(64, (3,3), activation='relu'),
MaxPooling2D((2,2)),
Flatten(),
Dense(128, activation='relu'),
Dense(10, activation='softmax')
])
cnn.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
h3 = cnn.fit(X_train_cnn, y_train_cat, epochs=5, validation_data=(X_test_cnn, y_test_cat))
```

```python
# Validation (Test) Accuracy Comparison Plot
plt.figure()
plt.plot(h1.history['val_accuracy'], label='Perceptron - Test')
plt.plot(h2.history['val_accuracy'], label='ANN - Test')
plt.plot(h3.history['val_accuracy'], label='CNN - Test')
plt.title("Validate Accuracy Comparison")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
#  Validation (Test) Loss Comparison Plot
plt.figure()
plt.plot(h1.history['val_loss'], label='Perceptron - Test')
plt.plot(h2.history['val_loss'], label='ANN - Test')
plt.plot(h3.history['val_loss'], label='CNN - Test')
plt.title("Validate Loss Comparison")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()
```

```python
#  Final Test Set Evaluation
p_test = perceptron.evaluate(X_test, y_test_cat, verbose=0)
a_test = ann.evaluate(X_test, y_test_cat, verbose=0)
c_test = cnn.evaluate(X_test_cnn, y_test_cat, verbose=0)


print("Final Test Results")
print(f"Perceptron -> Loss: {p_test[0]:.4f}, Accuracy: {p_test[1]*100:.2f}%")
print(f"ANN -> Loss: {a_test[0]:.4f}, Accuracy: {a_test[1]*100:.2f}%")
print(f"CNN -> Loss: {c_test[0]:.4f}, Accuracy: {c_test[1]*100:.2f}%")
plt.figure()
plt.plot(h1.history['accuracy'], label='Perceptron')
plt.plot(h2.history['accuracy'], label='ANN')
plt.plot(h3.history['accuracy'], label='CNN')
plt.title("Training Accuracy Comparison")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.show()
```

```python
#  Loss Comparison Plot
plt.figure()
plt.plot(h1.history['loss'], label='Perceptron')
plt.plot(h2.history['loss'], label='ANN')
plt.plot(h3.history['loss'], label='CNN')
plt.title("Training Loss Comparison")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.show()
```
